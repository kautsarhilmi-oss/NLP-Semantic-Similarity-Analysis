```python
import nltk
import string
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
sentences = [
    "Pemerintah mengumumkan kebijakan baru terkait ekonomi nasional",
    "Harga minyak dunia mengalami kenaikan signifikan",
    "Teknologi kecerdasan buatan berkembang dengan sangat pesat",
    "Sepak bola merupakan olahraga yang paling populer di dunia",
    "Ekonomi Indonesia menunjukkan pertumbuhan yang stabil",
    "Perkembangan teknologi digital mempengaruhi berbagai sektor industri",
    "Tim nasional berhasil memenangkan pertandingan dengan skor telak",
    "Penelitian terbaru menunjukkan peningkatan kualitas udara",
    "Pasar saham mengalami penurunan akibat kondisi global",
    "Inovasi dalam bidang teknologi semakin berkembang setiap tahun"
]
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')

def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = word_tokenize(text)
    stop_words = set(stopwords.words('indonesian'))
    words = [w for w in words if w not in stop_words]
    return ' '.join(words)

processed = [preprocess(s) for s in sentences]
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(processed)
```

```python
query = "teknologi"
query_vec = vectorizer.transform([query])

similarities = cosine_similarity(query_vec, X).flatten()
```

```python
print("Hasil Similarity:\n")

for i, score in enumerate(similarities):
    print(f"Kalimat {i+1}: {sentences[i]}")
    print(f"Similarity Score: {score:.4f}")
    print("-" * 40)
```
